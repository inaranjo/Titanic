```python
data["Embarked"] = data["Embarked"].fillna("S")
```

```python
data["Age"] = data["Age"].fillna(data.Age.mean())
```

```python
np.unique(data.Embarked)
```

```python
data['Embarked_id'] = np.zeros(data.shape[0])
```

```python
data.loc[data["Embarked"] == 'C','Embarked_id'] = 1
data.loc[data["Embarked"] == 'Q','Embarked_id'] = 2
data.loc[data["Embarked"] == 'S','Embarked_id'] = 3
```

```python
data.loc[data["sex_id"] == True,'sex_id'] = 1.0
data.loc[data["sex_id"] == False,'sex_id'] = 0.0
```

```python
data['Embarked_id']
```

```python
data.head()
```

```python
features = ['Age', 'SibSp', 'Parch', 'Fare', 'sex_id', 'Embarked_id','Pclass']
features
```

```python
X_train = data[features].values
Y_train = data.Survived.values
```

```python
type(Y_train)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf_RFC =  RandomForestClassifier(max_depth=5, random_state=0, n_estimators=200)
clf_RFC.fit(X_train,Y_train)
clf_GBC = GradientBoostingClassifier(max_depth=2, n_estimators=100, learning_rate=0.1)
clf_GBC.fit(X_train,Y_train)
```

```python
clf_RFC.feature_importances_
```

```python
clf_GBC.feature_importances_
```

```python
plt.hist(clf_RFC.predict_proba(X_train[data.Survived==0])[:,0], bins=100, range=[0,1])
plt.hist(clf_RFC.predict_proba(X_train[data.Survived==1])[:,0], bins=100, range=[0,1])
plt.show()
```

```python
plt.hist(clf_GBC.predict_proba(X_train[data.Survived==0])[:,0], bins=100, range=[0,1])
plt.hist(clf_GBC.predict_proba(X_train[data.Survived==1])[:,0], bins=100, range=[0,1])
plt.show()
```

```python
from sklearn.metrics import accuracy_score, auc
```

```python
accuracy_score(clf_RFC.predict(X_train), Y_train)
```

```python
accuracy_score(clf_GBC.predict(X_train), Y_train)
```

```python
from sklearn.metrics import roc_curve

y_pred_rfc = clf_RFC.predict_proba(X_train)[:, 1]
fpr_rfc, tpr_rfc, _ = roc_curve(Y_train, y_pred_rfc)

auc(fpr_rfc, tpr_rfc, reorder=True)
```

```python
y_pred_gbc = clf_GBC.predict_proba(X_train)[:, 1]
fpr_gbc, tpr_gbc, _ = roc_curve(Y_train, y_pred_gbc)

auc(fpr_gbc, tpr_gbc, reorder=True)
```

```python
plt.plot(fpr_rfc, tpr_rfc, label='RFC')
plt.plot(fpr_gbc, tpr_gbc, label='GBC')
plt.legend()
```
